```python
#Importacion de clases y librerias
import pandas as pd
```

```python
# Carga de dataset
df = pd.read_csv("..\\data\\raw\\ibm.clean.Education.csv")
df.sample(3)
```

```python
df.info()
```

```python
# Verificar si hay duplicados en el DataFrame
duplicates = df.duplicated()

# Contar el número de filas duplicadas
num_duplicates = duplicates.sum()

if num_duplicates > 0:
    # Eliminar los duplicados del DataFrame
    df_clean = df.drop_duplicates()
    
    print(f"Se encontraron {num_duplicates} filas duplicadas y fueron eliminadas.")
else:
    print("No se encontraron filas duplicadas en el DataFrame.")
```

```python
# Especifica las columnas que deseas considerar para la detección de duplicados
columns_to_check = ['MonthlyIncome', 'AgeStartedWorking', 'Age', 'TotalWorkingYears',
                     'YearsAtCompany', 'DistanceFromHome', 'YearsInCurrentRole',
                     'YearsWithCurrManager', 'YearsSinceLastPromotion','PercentSalaryHike']

# Verifica si hay duplicados considerando solo las columnas especificadas
duplicates_subset = df_clean.duplicated(subset=columns_to_check)

# Contar el número de filas duplicadas
num_duplicates_subset = duplicates_subset.sum()

print(f"Se encontraron {num_duplicates_subset} filas duplicadas considerando las columnas especificadas.")
```

```python
# Eliminar las filas duplicadas basadas en las columnas especificadas
df_clean = df_clean.drop_duplicates(subset=columns_to_check, keep='first')

# Verificar el tamaño del DataFrame después de eliminar las filas duplicadas
print("Tamaño del DataFrame después de eliminar duplicados:", df_clean.shape)
```
